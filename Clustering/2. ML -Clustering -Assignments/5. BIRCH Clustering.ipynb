```python
# import library

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load dataset

dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# x=unsupervised iloc-integer location

x=dataset.iloc[:,[3,4]].values
x
```

```python
# Finding number of clusters using dendrogram

import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Clusters")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
#Model creation with 4 clusters

from sklearn.cluster import Birch
brc = Birch(branching_factor = 50, n_clusters = 4, threshold = 1.5)
brc.fit(x)

labels = brc.labels_
print("cluser_group : \n", labels)
```

```python
# Add cluster group in existing dataset

supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=labels
supervised
```

```python
# save new cluster_group in new .csv file

supervised.to_csv("5. BIRCH Cluster_group.csv")
```

```python
# Output - sea born - Data visualization

import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4], hue=supervised.columns[5],
                fit_reg=False, legend=True, legend_out=True)
```
