```python
# import library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset

dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# (x,y) - Supervised
# x - Unsupervised
# [:,] - include all rows
# [3:4] - include 3rd and 4th column
# iloc - integer location
# iloc - integer based indexing
# iloc[: or ,]

# x=dataset.iloc[:,[3,4]].values
# , or :
x=dataset.iloc[:,3:5].values
x
```

```python
# Finding number of clusters in a dataset using Elbow method
#.inertia_ - is calculated by measuring the distance between each data point and its centroid
# WCSS - It is defined as the sum of square distances between each data point and its centroid.
# WCSS - (Within-Cluster Sum of Square)

# sklearn kmeans clustering

# n_clustersint, default=8
# init{â€˜k-means++â€™, â€˜randomâ€™}, callable or array-like of shape (n_clusters, n_features), default=â€™k-means++â€™

from sklearn.cluster import KMeans
list1=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    list1.append(kmeans.inertia_)
plt.plot(range(1,11),list1)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Model creation using 5 clusters

from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
ykmeans=kmeans.fit_predict(x)
ykmeans
```

```python
#  Centroids for 5 clusters

centroids=kmeans.cluster_centers_
centroids
```

```python
# Add cluster_group in existing dataset

supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=ykmeans
supervised
```

```python
# save cluster_group in new .csv file

# .to_csv = allows you to save a Pandas DataFrame as a CSV file
# converts DataFrame into CSV data

supervised.to_csv("1. KMeans_Cluster_group.csv")
```
