```python
#import pandas library

import pandas as pd
```

```python
# load datset

dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
# Categorical Input data

dataset=pd.get_dummies(dataset, drop_first=True)
dataset
```

```python
# view dataset column names

dataset.columns
```

```python
# split input data

independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
independent
```

```python
# split output data

dependent=dataset[['Profit']]
dependent
```

```python
# Split training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Model Creation - Decision Tree

# Hyper Tuning Parameter
# DecisionTreeRegressor() - no parameter - default parameter-squared error

# Parameters:
# criterion{“squared_error”, “friedman_mse”, “absolute_error”, “poisson”}, default=”squared_error”

# splitter{“best”, “random”}, default=”best”

from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error',splitter='best')

regressor=regressor.fit(x_train, y_train)
```

```python
# Decision Tree - View Graph 
# Matplotlib is used for data visualization
# Matplotlib is a plotting library in python

import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluation metrics

y_pred=regressor.predict(x_test)
```

```python
# r2 value

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
